```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
data = yf.download("MSFT",start="2021-05-23", end="2023-05-23")
hp , lp , cp , v = data['High'] , data['Low'] , data['Close'] , data['Volume']
l=len(data)
RoC , NVIs , PSAR = [0]*l , [0]*l , [lp[0]]*l
ep = hp[0]
afstr, afinc, afmax = 0.02, 0.02, 0.2
af , dir = afstr , 1
# As we consider a rounded-off value on the starting date
NVIs[0] = 1000
```

```python
for i in range(1,l):
    # Calculating RoC
    RoC[i] = ((((cp[i] - cp[0])/cp[0])))*100

    # Calculating Negative Volume Indices
    if v[i]<v[i-1]:
        NVIs[i] = NVIs[i-1] + ((cp[i]-cp[i-1])/cp[i-1])*NVIs[i-1]
    else:
        NVIs[i]=NVIs[i-1]

    # Calculating Parabolic SAR
    if dir == 1:
        if lp[i] < PSAR[i-1]:
            dir = -1
            PSAR[i] = ep
            ep = hp[i]
            af = afstr
        else:
            PSAR[i] = PSAR[i-1] + af*(ep - PSAR[i-1])
            ep = max(ep,hp[i])
            af = min(af + afinc , afmax)
    else:
        if hp[i] > PSAR[i-1]:
            dir = 1
            PSAR[i] = ep
            ep = lp[i]
            af = afstr
        else:
            PSAR[i] = PSAR[i-1] + af*(ep-PSAR[i-1])
            ep = min(ep,lp[i])
            af = min(af + afinc , afmax)
```
